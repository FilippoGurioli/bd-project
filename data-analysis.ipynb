# Data analysis of results

This document presents a comprehensive analysis of the results obtained from the experiment.

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_json("results/optimized.json", lines=True)
print(df.head())
print("\nColumns:", df.columns.tolist())
df_sorted = df.sort_values("avg_tip_pct", ascending=False)
```

## Top 10 zones by average tip percentage

```python
top10 = df_sorted.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top10["Zone"], top10["avg_tip_pct"], color="skyblue")
plt.xlabel("Average Tip Percentage")
plt.title("Top 10 NYC Taxi Zones by Average Tip %")
plt.gca().invert_yaxis()  # Highest on top
plt.tight_layout()
plt.show()
```

## Average tip percentage by borough

```python
plt.figure(figsize=(8, 5))
borough_avg = df.groupby("Borough")["avg_tip_pct"].mean().sort_values(ascending=False)
borough_avg.plot(kind="bar", color="orange")
plt.ylabel("Average Tip %")
plt.title("Average Tip Percentage by Borough")
plt.tight_layout()
plt.show()
```

## Distribution of average tip percentages

```python
plt.figure(figsize=(8, 5))
plt.hist(df["avg_tip_pct"], bins=20, color="lightgreen", edgecolor="black")
plt.xlabel("Average Tip Percentage")
plt.ylabel("Frequency")
plt.title("Distribution of Avg Tip % across Zones")
plt.tight_layout()
plt.show()
```
